```
import tensorflow as tf
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
```

```
def rbf(x,c,s):
    return np.exp(-np.sum((x-c)**2, axis=1)/(2*s**2))
```

```
mnist = tf.keras.datasets.mnist
```

```
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train = tf.reshape(x_train, [60000, 784])
x_test = tf.reshape(x_test, [10000, 784])

x_train = tf.Session().run(x_train)
x_test = tf.Session().run(x_test)
y_train.shape
```

```
df = pd.DataFrame()
temp = {}
```

```
sd = [1,2,4,6,8,10,12,16]
epoch = [100,200,400,500,600,800]
ac = []
ls = []

for m in epoch:
    kmeans = KMeans(n_clusters=1024,n_init=1,max_iter=10, random_state=0).fit(x_train[:10000,:784])
    W = kmeans.cluster_centers_
    print("FOR EPOCH = ",m)
    for i in tqdm(range(len(W))):
      df[i] = rbf(x_train[:10000,:784],W[i],2)
      temp['data']= df
      temp['weights']= W


    data = temp['data']
    data['target'] = y_train[:10000,]
    data.head()


    from sklearn.model_selection import train_test_split
    X = data.drop('target',1).values
    Y = data['target']
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)

    X_train = X_train / np.linalg.norm(X_train, axis=1).reshape((8000,1))
    X_test = X_test / np.linalg.norm(X_test, axis=1).reshape((2000,1))

    Y_train = pd.get_dummies(Y_train).values
    Y_test = pd.get_dummies(Y_test).values



    x = tf.placeholder(tf.float64,shape=(None,1024))
    y_true = tf.placeholder(tf.float64,shape=(None,10))
    W = tf.Variable(tf.zeros([x.shape[1], 10],tf.float64))
    b = tf.Variable(tf.zeros([10], dtype=tf.float64))
    logits = tf.matmul(x,W) + b
    y_pred = tf.nn.softmax(logits)


    lr = 0.01
    epochs = m
    cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_true))


    loss_history = []
    accuracy_score = []
    training = tf.train.AdamOptimizer(lr).minimize(cross_entropy)
    init = tf.global_variables_initializer()
    with tf.Session() as sess:
        sess.run(init)
        for i in range(epochs+1):
            sess.run(training, feed_dict={x: X_train, y_true: Y_train})
            correct_prediction = tf.equal(tf.argmax(y_pred,1), tf.argmax(y_true,1))
            accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float64))
            accuracy_score.append(sess.run(accuracy, feed_dict={x: X_test, y_true: Y_test}))
            loss_history.append(sess.run(cross_entropy, {x: X_train, y_true: Y_train}))
            if i%10 == 0:
                print('Step:' + str(i) + '  Accuracy =  ' +
                  str(100*sess.run(accuracy, feed_dict={x: X_test, y_true: Y_test})) +
                  '  Loss = ' + str(sess.run(cross_entropy, feed_dict={x: X_train, y_true: Y_train})))


    plt.plot(loss_history, label='Loss value')
    plt.title('Training Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()


    plt.plot(accuracy_score, label='accuracy')
    plt.title('Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('accuracies')
    plt.show()

    a = np.array(accuracy_score)
    l = np.array(loss_history)
    ac.append(np.mean(a))
    ls.append(np.mean(l))
```

```
plt.plot(epoch,ls)
plt.title('LOSS VS EPOCH')
plt.xlabel('EPOCH')
plt.ylabel('LOSS')
plt.show()

plt.plot(epoch,ac)
plt.title('ACC VS EPOCH')
plt.xlabel('EPOCH')
plt.ylabel('ACC')
plt.show()
```
